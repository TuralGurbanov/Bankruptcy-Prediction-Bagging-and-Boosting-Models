# Predicting whether companies will go bankrupt or not using:
- Bagging Classifier
- Gradient Boosting Classifier
- XGBoost Classifier
- Catboost Classifier

```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv("data.csv")
df = data.copy()
df.head()
```

## Exploratory Data Analysis and Data Preparation

```python
df.isnull().sum().sum()
```

```python
df.describe()
```

```python
df.nunique()
```

```python
unique_columns = []
for col in df.columns:
    if(df[col].nunique()<20):
        unique_columns = unique_columns+[col]
```

```python
unique_columns.remove('Bankrupt?')
unique_columns
```

```python
df.drop(columns=unique_columns).head()
```

```python
df_dummy = pd.get_dummies(df.loc[:, unique_columns])
df_dummy.isnull().sum().sum()
```

```python
df.drop(columns=unique_columns).isnull().sum().sum()
```

```python
df_new = pd.concat([df.drop(columns=unique_columns),df_dummy], axis = 1)
```

```python
df_new
```

```python
y = df_new.iloc[:, 0]
X = df_new.drop(columns = 'Bankrupt?')
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 123, stratify = y)
```

```python
X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

## Bagging Classifier

```python
from sklearn.ensemble import BaggingClassifier
from sklearn import tree
model = BaggingClassifier(tree.DecisionTreeClassifier(random_state=1))
model.fit(X_train, y_train)
model.score(X_test, y_test)
```

## Gradient Boosting Classifier

```python
from sklearn.ensemble import GradientBoostingClassifier
model = GradientBoostingClassifier(learning_rate = 0.01, random_state=1)
model.fit(X_train, y_train)
model.score(X_test, y_test)
```

## XGBoost Classifier

```python
from xgboost import XGBClassifier
```

```python
xgbt = XGBClassifier(max_depth = 2,
             learning_rate = 0.2,
             objective  = "multi:softmax",
             num_class = 2,
             booster = "gbtree",
             n_estimarors = 10,
             random_state = 123)
```

```python
xgbt.fit(X_train, y_train)
```

```python
xgbt_pred = xgbt.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(y_test, xgbt_pred)
```

```python
xgbt.score(X_train, y_train)
```

```python
xgbt.score(X_test, y_test)
```

## Catboost

```python
from catboost import CatBoostClassifier

# Initialize CatBoostRegressor
model = CatBoostClassifier(iterations=2,
                          learning_rate=1,
                          depth=2)
# Fit model
model.fit(X_train, y_train)
# Get predictions
preds = model.predict(X_test)
```

```python
model.score(X_train, y_train)
```

```python
model.score(X_test, y_test)
```

```python
accuracy_score(y_test, preds)
```
